```python
import pandas as pd
train_data = pd.read_csv("train_QC_data.csv")
val_data = pd.read_csv("val_QC_data.csv")
test_data = pd.read_csv("test_QC_data.csv")
```

```python
train_data
```

```python
final_data = pd.concat([train_data,val_data,test_data])
final_data
```

Get columns and split data

```python
final_data_trimmed = final_data[["Question","QCLabel"]]
```

```python
final_data_trimmed
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(final_data_trimmed["Question"].values)
```

```python
vectorizer.get_feature_names_out()[300:400]
```

```python
features.shape
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
labels = le.fit_transform(final_data_trimmed["QCLabel"].values)
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,labels, test_size=0.33, random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train,y_train)
labels_test = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,labels_test)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
labels = le.fit_transform(final_data_trimmed["QCLabel"].values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_data_trimmed["Question"].values,labels, test_size=0.33, random_state=42)
```

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(X_train)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(features,y_train)
X_test = vectorizer.transform(X_test)
labels_test = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,labels_test)
```

# dimensionality reduction

```python
features.shape
```

```python
from sklearn.decomposition import PCA
pca_transform = PCA(n_components=1000)

X_reduced = pca_transform.fit_transform(features)
X_reduced.shape
```

```python
from sklearn.decomposition import TruncatedSVD
trunc_svd = TruncatedSVD(n_components=1000, n_iter=7, random_state=42)

X_reduced = trunc_svd.fit_transform(features)
X_reduced.shape
```
